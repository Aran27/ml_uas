```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = './A2022_dataset_buat bedah data.csv'
# path ke dataset, lebih baik dalam satu folder

df = pd.read_csv(path, on_bad_lines='skip', delimiter=';')
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

```python
df['namapemda'].unique()
```

```python
# kota/kab kalimantan utara = kota
kota = df[df['kodepemda']>34]
kota = kota.dropna(axis=1,how='all')
kota.head()
```

```python
kota[['nilaianggaran']] = kota[['nilaianggaran']].fillna(method='ffill')
kota.fillna(value='-',inplace=True)

# melihat kolom yang masih ada nan
kota.columns[kota.isna().any()]
```

```python
kota.head()
```

```python
sub1_kota = kota.groupby(['namapemda']).sum()
sub1_kota
```

```python
sub1_kota.plot.barh(y='nilaianggaran', rot = 0, figsize=(8,5))
plt.ylabel('Nama Pemerintahan Daerah')
plt.xlabel('Jumlah Anggaran (Triliun)')
plt.title('Anggaran Kota/Kabupaten di Kalimantan Utara')
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(kota['nilaianggaran'])
```

```python
def detect_outliers(df, x):
    Q1 = df[x].describe()['25%']
    Q3 = df[x].describe()['75%']
    IQR = Q3-Q1
    return df[(df[x] < Q1-1.5*IQR) | (df[x] > Q3+1.5*IQR)]

sub2_kota = kota
```

```python
# penghapusan outlier
ind1 = [0]
while ind1.__len__() != 0:
    df2 = detect_outliers(sub2_kota,'nilaianggaran')
    ind1 = df2.index
    sub2_kota = sub2_kota.drop(index=ind1)
```

```python
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']

newdf = sub2_kota.select_dtypes(include=numerics)
newdf.head()
```

```python
from sklearn.cluster import KMeans
cs = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(newdf)
    cs.append(kmeans.inertia_)
plt.plot(range(1, 11), cs)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('CS')
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4,random_state=0)

kmeans.fit(newdf)

labels = kmeans.labels_
labels

```

```python
#Hasil cluster adalah kolom cluster

newdf['cluster']=labels
newdf.head()
```
